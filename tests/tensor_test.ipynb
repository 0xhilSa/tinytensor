```python
import tinytensor as tt
```

```python
tt.__version__
```

```python
x = tt.Tensor([tt.pi/4 * i for i in range(10)], dtype=tt.float16)
x
```

```python
print(f"sin: {x.sin().data()}")
print(f"cos: {x.cos().data()}")
print(f"tan: {x.tan().data()}")
print(f"sinh: {x.sinh().data()}")
print(f"cosh: {x.cosh().data()}")
print(f"tanh: {x.tanh().data()}")
```

```python
cuda_x = x.cuda()
cuda_x
```

```python
print(f"sin: {cuda_x.sin().data()}")
print(f"cos: {cuda_x.cos().data()}")
print(f"tan: {cuda_x.tan().data()}")
print(f"sinh: {cuda_x.sinh().data()}")
print(f"cosh: {cuda_x.cosh().data()}")
print(f"tanh: {cuda_x.tanh().data()}")
```

```python
cuda_x.sin() + cuda_x.tan()
```

```python
(cuda_x.sin() + cuda_x.tan()).data()
```
