```python
import tinytensor as tt
import torch
```

```python
x = tt.Tensor([1,2,3,4,5], dtype=tt.uint64, device="cuda")
y = tt.Tensor([2,3,4,5,6], dtype=tt.int8, device="cuda")
tx = torch.tensor([1,2,3,4,5])
ty = torch.tensor([2,3,4,5,6])
```

```python
x.bitwise_and(y)
```

```python
x.bitwise_and(y).data(), tx.bitwise_and(ty)
```

```python
x.bitwise_or(y).data(), tx.bitwise_or(ty)
```

```python
x.bitwise_xor(y).data(), tx.bitwise_xor(ty)
```

```python
x.bitwise_nand(y).data(), torch.bitwise_not(tx.bitwise_and(ty))
```

```python
x.bitwise_nor(y).data(), torch.bitwise_not(tx.bitwise_or(ty))
```

```python
x.bitwise_xnor(y).data(), torch.bitwise_not(tx.bitwise_xor(ty))
```

```python
x.bitwise_not().data(), tx.bitwise_not()
```

```python
y.bitwise_not().data(), ty.bitwise_not()
```

```python
x.logical_and(y).data(), tx.logical_and(ty)
```

```python
x.logical_and(y).data(), tx.logical_and(ty)
```
